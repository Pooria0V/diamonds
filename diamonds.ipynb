```python
import pandas as pd
import numpy as np
from tensorflow import keras
import tensorflow as tf
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
missing_data = df.isnull().sum()
missing_data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['color'] = le.fit_transform(df['color'])
df['cut'] = le.fit_transform(df['cut'])
df['clarity'] = le.fit_transform(df['clarity'])
df.drop('Unnamed: 0', axis=1, inplace=True)

df.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming 'data' is your dataset
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
# scaled_data
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('price', axis=1)  # Features
y = df['price']               # Target variable
x_train_main, x_test_main, y_train_main, y_test_main = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error


model = keras.Sequential()
model.add(keras.layers.Input(shape=(9,)))
model.add(keras.layers.Dense(units=32, activation='relu'))
model.add(keras.layers.Dense(units=32, activation='sigmoid'))
model.add(keras.layers.Dense(units=1))
model.compile(optimizer=tf.optimizers.Adam(), loss='mean_absolute_error', metrics=['accuracy'])
model.fit(x_train_main, y_train_main, epochs=20, batch_size=64, validation_data=(x_test_main, y_test_main))
y_pred = model.predict(x_test_main)
mse = mean_absolute_error(y_true=y_test_main, y_pred=y_pred)
mse
```
